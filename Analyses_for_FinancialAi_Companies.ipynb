# How much the companies spent on R&D

```python
df.head()
```

```python
print("Company spending in $Bn")
RD = df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000
RD
```

```python
plt.bar(RD.index, RD.values, color = ['cyan', 'yellow', 'grey'])
plt.title("R&D Spending by the Companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")
plt.show();
```

Revenue Earned by the Companies


```python
df.head()
```

```python
print("Company's AI Revenue USD_$Bn")
rev = df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000
rev
```

```python
plt.bar(rev.index, rev.values, color = ['cyan', 'yellow', 'grey'], width = 0.4)
plt.title("AI Revenue by the Companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")
plt.show();
```

Bar Plots to show Expenditure and Revenues of the companies

```python
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.bar(rev.index, rev.values, color = ['cyan', 'yellow', 'grey'], width = 0.4)
plt.title("R&D spending the Companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.subplot(1, 2, 2)
plt.bar(rev.index, rev.values, color = ['cyan', 'yellow', 'grey'], width = 0.4)
plt.title("AI Revenue by the Companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")
plt.show();
```

Expenditure Vs Revenue Year-by-Year

```python
df.head()
```

```python
df['Year'] = df['Date'].dt.year
spend = df.groupby('Year')['R&D_Spending_USD_Mn'].sum()
spend
```

```python
plt.plot(spend.index,  spend.values, color = 'red');
plt.title(" Combined R&D Spending Year-by-Year")
plt.xlabel("Year")
plt.ylabel("Amount in USD_$Mn")
plt.show()
```

```python
revenue = df.groupby('Year')['AI_Revenue_USD_Mn'].sum()
revenue
```

```python
plt.plot(revenue.index, revenue.values, color = 'green')
plt.title("Combined AI Revenue Year-by-Year")
plt.xlabel("Year")
plt.ylabel("Amount in USD_$Mn")
plt.show()
```

```python
plt.plot(spend.index,  spend.values, color = 'red');
plt.plot(revenue.index, revenue.values, color = 'green')
plt.title("Combined R&D Expenditure and AI Revenue Year-by-Year")
plt.xlabel("Year")
plt.ylabel("Amount in USD_$Mn")
plt.legend(['Expenditure',  'Revenue'])
plt.show()
```

```python
dept = df.groupby('Department')['Salary_INR'].mean()/1000
dept
```

```python
type(dept)
```

```python
dept.plot(x = dept.index, y = dept.values, kind = 'bar', color = 'cyan', legend=True, width=0.3);
plt.grid()
plt.title('Average Salary per Department')
plt.ylabel('Salary');
```

```python
df.head(2)
```

```python
salary = df.groupby('Job_Title')['Salary_INR'].mean()/1000
salary

```

```python
salary.plot(x=salary.index, y=salary.values, kind='bar', color='g',legend=True, width=0.3);
plt.grid(True)
plt.title('Average Salary per Job Title')
plt.ylabel('Salary');
```

```python
dept_job = df.groupby(['Job_Title','Department'])['Salary_INR'].mean()/1000
dept_job
```

```python


dept_job.plot(kind = 'barh',figsize=(10,6));
plt.title('Average Salary per Job Title and Department')
plt.xlabel('Salary')
plt.savefig('chart.png');
```

```python
df_Resigned = df[df['Status'] == 'Resigned']
df_Resigned

```

```python
Emp_terminated = df[df['Status'] == 'Terminated']
Emp_terminated
```

```python
terminated_emp_count = Emp_terminated.groupby('Department')['Status'].count()
terminated_emp_count.plot(kind='bar', color='orange', legend=True, width=0.3)
plt.title('Terminated Employees per Department')
plt.ylabel('Count');
```
